# Then, import cleaned data from yahoo api:

```python
btc = pd.read_csv('./4_yahoo_api_data/final_btc.csv',header=0)
btc=pd.DataFrame(btc)
btc= btc[['Close','dt']]
new_cols = ['bitcoin_usd','timestamp']
btc.columns = new_cols
btc
```

```python
eth = pd.read_csv('./4_yahoo_api_data/final_eth.csv',header=0)
eth=pd.DataFrame(eth)
eth= eth[['Close','dt']]
new_cols = ['ethereum_usd','timestamp']
eth.columns = new_cols
eth
```

```python
doge = pd.read_csv('./4_yahoo_api_data/final_doge.csv',header=0)
doge=pd.DataFrame(doge)
doge= doge[['Close','dt']]
new_cols = ['doge_usd','timestamp']
doge.columns = new_cols
doge
```

```python
NZX50 = pd.read_csv('./4_yahoo_api_data/final_NZindex.csv',header=0)
NZX50 =pd.DataFrame(NZX50)
NZX50 = NZX50[['Close','dt']]
new_cols = ['NZX50_index','timestamp']
NZX50.columns = new_cols
NZX50
```

```python
# final step: merge 5 dataframes into a big dataframe:
df_sentiment= pd.read_csv('./5_merge_data/final_sentiment.csv')
```

```python
df_sentiment['timestamp'] = pd.to_datetime(df_sentiment['timestamp'])
df_sentiment
```

```python
#get the sentiment score for 5/3 to 5/6
df_sentiment= df_sentiment[['sentiment_score','timestamp']]
df_sentiment
```

```python
# start to merge from api data first
all_api = pd.merge(NZX50,btc, on='timestamp')
all_api =pd.merge(all_api, eth, on='timestamp')
all_api =pd.merge(all_api,doge, on='timestamp')
all_api
```

```python
all_api['timestamp'] = pd.to_datetime(all_api['timestamp'])
```

```python
#final data merged (api data with sentiment score)
all_df = pd.merge(df_sentiment,all_api,on='timestamp')
all_df
```

```python
#export the final data
all_df.to_csv('./5_merge_data/final_all_data.csv')
```
